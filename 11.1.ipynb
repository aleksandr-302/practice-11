```python
from random import randint
def bubble(n):
  lst = []
  temp=0
  for i in range(n):
    lst.append(randint(1, 100))
  print(lst)
 
  for i in range(0, n-1):
      for j in range(0, n-i-1):
          if lst[j] > lst[j+1]:
              temp=lst[j+1]
              lst[j+1]=lst[j]
              lst[j]=temp
  print("отсортированный: ",lst)
  print('____')
  
bubble(10)
bubble(100)
```
